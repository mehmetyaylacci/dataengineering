```python
import pandas as pd

df = pd.read_csv('./transfermarkt/clubs.csv', encoding='UTF-8')
```

```python
df.head()
```

```python
df = df.drop(columns=['total_market_value'])
df.head()
```

```python
df = df.drop(columns=['club_code','squad_size','average_age','foreigners_number','foreigners_percentage','national_team_players','stadium_seats','url'])
df.head()
```

```python
df_small = df[(df.domestic_competition_id == "FR1") | (df.domestic_competition_id == "TR1") | (df.domestic_competition_id == "GB1") | (df.domestic_competition_id == "IT1") | (df.domestic_competition_id == "L1") | (df.domestic_competition_id == "ES1")]
df_small.head()
```

```python
df_small = df_small.drop(columns="coach_name")
```

```python
df_small.to_csv('./transfermarkt_cleaned/clubs.csv', encoding="UTF-8", header=True, index=False)
```

This part is going to be players

```python
df_clubs = df_small

df_players = pd.read_csv('./transfermarkt/players.csv', encoding='UTF-8')
df_players.head()
```

```python
df_players = df_players[df_players['current_club_id'].isin(df_clubs['club_id'])]
df_players.head()
```

```python
df_players = df_players[['player_id','name','last_season','current_club_id','height_in_cm','market_value_in_eur','highest_market_value_in_eur']]
df_players.head()
```

```python
df_players = df_players[df_players['last_season'] == 2022]
```

```python
df_players.to_csv('./transfermarkt_cleaned/players.csv', header=True, encoding='UTF-8', index=False)
```

Player valuations part

```python
df_valuations = pd.read_csv('./transfermarkt/player_valuations.csv', encoding='UTF-8')
df_valuations.head()
```

```python
df_valuations = df_valuations[['player_id','last_season','date','market_value_in_eur','current_club_id']]
df_valuations.head()
```

```python
df_valuations = df_valuations[df_valuations['player_id'].isin(df_players['player_id'])]
df_valuations.head()
```

```python
df_valuations.to_csv('./transfermarkt_cleaned/valuation.csv', header=True, encoding='UTF-8', index=False)
```

Now the other parts

```python
df_games = pd.read_csv('./transfermarkt/games.csv', encoding='UTF-8')
df_games.head()
```

```python
df_games = df_games[(df_games['home_club_id'].isin(df_clubs['club_id'])) | (df_games['away_club_id'].isin(df_clubs['club_id']))]
df_games.head()
```

```python
df_games.to_csv('./transfermarkt_cleaned/games.csv', header=True, encoding='UTF-8', index=False)
```

```python
df_appr = pd.read_csv('./transfermarkt/appearances.csv', encoding='UTF-8')
df_appr.head()
```

```python
df_appr = df_appr[df_appr['player_id'].isin(df_players['player_id'])]
df_appr.head()
```

```python
df_appr.to_csv('./transfermarkt_cleaned/appearances.csv', header=True, encoding='UTF-8', index=False)
```

```python
df_appr['date'] = pd.to_datetime(df_appr['date'])
df_appr_filtered = df_appr[df_appr['date'] > '2023-01-01']
df_appr_filtered.head()
```

```python
df_appr_filtered.to_csv('./transfermarkt_cleaned/appearances.csv', header=True, encoding='UTF-8', index=False)
```

```python
df_games = df_games[df_games['game_id'].isin(df_appr_filtered['game_id'])]
```

```python
df_games.to_csv('./transfermarkt_cleaned/games.csv', header=True, encoding='UTF-8', index=False)
```

```python
df_valuations = df_valuations[df_valuations['player_id'].isin(df_players['player_id'])]
df_valuations.head()
```

```python
df_valuations.to_csv('./transfermarkt_cleaned/valuations.csv', header=True, encoding='UTF-8', index=False)
```
